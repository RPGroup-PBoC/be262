# Euler's method for numerical integration

```python
# For numerical computations
import numpy as np

# For plotting
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()  # Set seaborn's default style for the notebook

# This enables high resolution graphics inline
%config InlineBackend.figure_format = 'retina'
```

We will now numerically integrate the equation for cell growth of the form
$$
N(t  + \Delta t) = N(t) + (k \Delta t) N(t),
$$
where bla bla bla.

```python
# Define parameters for our differential equation
N_0 = 1  # initial number of cells
k = 0.03  # growth rate (min^-1)

# Define parameters for our integration
dt = 0.1  # min
total_time = 120  # min
```

Having defined the parameters we are ready to integrate the equation.

```python
# Determine number of time steps
num_steps = int(total_time / dt)

# Initialize array of number of cells
N_t = np.zeros(num_steps)

# Modify first entry to be initial number of cells
N_t[0] = N_0

# Use a for loop to compute number of cells over time
for i in np.arange(1, num_steps):
    # Apply the recipe
    N_t[i] = N_t[i-1] + k * dt * N_t[i-1]
```

Now that we integrated the equation let's plot it!

```python
# Define time array
time = np.arange(num_steps) * dt

# Plot growth curve
plt.plot(time, N_t, linewidth=2)

# Label axis
plt.xlabel('time (min)')
plt.ylabel(r'$N(t)$')
```

Let's now compare numerical integration to analytical solution of the ODE.

```python
# Compute closed-form solution
N_analytic = N_0 * np.exp(k * time)

# Plot growth curve
plt.plot(time, N_t, linewidth=2,
         label='numerical integration')
plt.plot(time, N_analytic, linewidth=2,
         label='analytical solution')

# Label axis
plt.xlabel('time (min)')
plt.ylabel(r'$N(t)$')
# Add legend
plt.legend()
```
