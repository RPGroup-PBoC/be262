# Study of the diffusion timescale

In the [previous tutorial](diffusion_via_coin_flips.html), through numerical simulation we noticed a linear relation between the variance of the diffusion distance and the diffusion time, namely, $L^2=D T$, where $L$ is the diffusion lengthscale and $T$ is the diffusion timescale. The coefficient of proportionality ($D$) in this relation is called the diffusion constant and has units of distance$^2$/time.

In this tutorial, we will numerically explore the $L^2=D T$ relation to get an intuition about the different scaling behaviors in diffusive processes.

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt

# Show figures in the notebook
%matplotlib inline

# for nice plots
import seaborn as sns
rc={'lines.linewidth': 2, 'axes.labelsize': 14, 'axes.titlesize': 14, \
    'xtick.labelsize' : 14, 'ytick.labelsize' : 14}
sns.set(rc=rc)
```

We choose a typical value for the diffusion constant inside a cell ($D = 10 \, \mu^2/\text{sec}$) and estimate the diffusion time for a range of different lengthscales.

```python
# diffusion constant in µm^2/sec, typical for protein 
D = 10

# range of lengths in microns from 10^0 to 10^6
L = np.logspace(0, 6, 100)

# calculate time to diffuse
t = L**2 / D
```

In the above chunk of code we used `logspace` to generate 100 points uniformly spread in the logarithmic scale between $10^0 = 1 \, \mu \text{m}$ and $10^6 \, \mu \text{m} = 1 \, \text{m}$.

Let's plot the relation.

```python
# Mmke a plot in linear scale
plt.plot(L, t)
plt.xlabel('length (µm)')
plt.ylabel('time (sec)')
```

As we can see from the plot, it would take about $10^{11} \, \text{sec} \sim 3000$ years to diffuse for 1 meter. This means that diffusion is effective only at small lengthscales. The linear scale of the plots does not allow us to explore small lengthscales. Let's redo the plot, this this in a logarithmic scale.

```python
# Make a log-log plot of the diffusion timescale
plt.loglog(L, t)
plt.xlabel('length (µm)')
plt.ylabel('time (sec)')
```

For distances of several microns, typical of bacteria, the diffusion timescale is on the order of a second. That's why diffusion is widely used by bacterial cells as means of transport. As we go to lengthscales of hundreds of microns, the diffusion times become on the order of minutes to an hour. That's why in eukaryotic cells directed transport is often used to bridge two parts of the cell, since by diffusion only the transport would be too slow.
