# Modeling diffusion via coin flips

This exercise was generated from a Jupyter notebook. You can download the notebook [here](diffusion_via_coin_flips_in_class.ipynb).

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt

# Show figures in the notebook
%matplotlib inline

# For pretty plots
import seaborn as sns
sns.set()
```

```python
# Random number in (0,1)
np.random.rand()
```

```python
# Number of steps
n_steps = 1000

# Array to store walker positions
positions = np.zeros(n_steps)

# Initialize the walker's position
positions[0] = 0

for i in range(n_steps-1):
    
    p = np.random.rand()
    
    if p > 0.5:
        positions[i+1] = positions[i] + 1
    else:
        positions[i+1] = positions[i] - 1

# Show the trejectory
plt.plot(positions)
plt.xlabel('Step')
plt.ylabel('Position')
plt.show()
```

```python
# Number of trajectories
n_traj = 1000

# 2d array for storing the trajectories
positions_2d = np.zeros([n_traj, n_steps])

# Initialize the position of all the walkers to 0
positions_2d[:,0] = 0

# Generate trajectories
for i in range(n_steps-1):
    for t in range(n_traj):
        p = np.random.rand()
        if p > 0.5:
            positions_2d[t,i+1] = positions_2d[t,i] + 1
        else:
            positions_2d[t,i+1] = positions_2d[t,i] - 1
```

```python
for t in range(n_traj):
    plt.plot(positions_2d[t,:])
plt.xlabel('Step')
plt.ylabel('Position')
plt.show()
```

```python
for t in range(n_traj):
    # alpha: transparency
    plt.plot(positions_2d[t,:], alpha = 0.01, color = 'black')
plt.xlabel('Step')
plt.ylabel('Position')
plt.show()
```

```python
# Make a histogram of final positions
_ = plt.hist(positions_2d[:,n_steps-1], bins = 20)
plt.xlabel('Final position')
plt.ylabel('Count')
```

```python
# For storing the variance
variance = np.zeros(n_steps)

# Calculate the variance
for i in range(n_steps):
    variance[i] = np.var(positions_2d[:,i])
    
# Plot the variance
plt.plot(variance)
plt.xlabel('Step')
plt.ylabel('Variance')
plt.show()
```
